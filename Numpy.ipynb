Continuando con el proyecto de las naranjas/toronjas, ahora debes calcular el coeficiente angular y lineal para la recta de las naranjas y para la recta de las toronjas. Utiliza la fórmula de mínimos cuadrados para encontrar cada uno.



Sabemos que:

Naranjas → Filas 0 a 4999 en datos_a

Toronjas → Filas 5000 en adelante en datos_a

Diámetro está en la columna 0

Peso está en la columna 1

```python
#DEFINIR DATOS

# Extraer datos de naranjas
X_naranjas = datos_a[:5000, 0]  # Diámetro
Y_naranjas = datos_a[:5000, 1]  # Peso

# Extraer datos de toronjas
X_toronjas = datos_a[5000:, 0]  # Diámetro
Y_toronjas = datos_a[5000:, 1]  # Peso
```

```python
def calcular_pendiente(X, Y):
    n = len(X)
    beta_1 = (np.sum(X * Y) - n * np.mean(X) * np.mean(Y)) / (np.sum(X**2) - n * (np.mean(X)**2))
    return beta_1

# Calcular pendiente para naranjas y toronjas
beta1_naranjas = calcular_pendiente(X_naranjas, Y_naranjas)
beta1_toronjas = calcular_pendiente(X_toronjas, Y_toronjas)

print(f"Pendiente (β₁) de naranjas: {beta1_naranjas:.4f}")
print(f"Pendiente (β₁) de toronjas: {beta1_toronjas:.4f}")
```
